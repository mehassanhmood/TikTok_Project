```python
# import Libraries
import pandas as pd
import numpy as np
```

```python
# Accessing data
data = pd.read_csv('tiktok_dataset.csv')
```

```python
data.head(2)
```

```python
# Information summary about the data set 
data.info()
```

```python
# getting descriptive statistics
data.describe()
```

```python
# Finding the average view count of videos with "claim" status
claims = data[data['claim_status'] == 'claim']
print('Mean view count claims:', claims['video_view_count'].mean())
print('Median view count claims:', claims['video_view_count'].median())
```

```python
# Finding the average view count of videos with "opinion" status
opinions = data[data['claim_status'] == 'opinion']
print('Mean view count opinions:', opinions['video_view_count'].mean())
print('Median view count opinions:', opinions['video_view_count'].median())
```

```python
# Determining the different values for claim status and how many of each are in the data:
data['claim_status'].value_counts()
```

```python
# Getting counts for each group combination of claim status and author ban status
data.groupby(['claim_status', 'author_ban_status']).count()[['#']]
```

```python
# Investigating data about author_ban_status
data.groupby(['author_ban_status']).agg(
    {'video_view_count': ['mean', 'median'],
     'video_like_count': ['mean', 'median'],
     'video_share_count': ['mean', 'median']})
```

```python
# Getting the median video share count of each author ban status
data.groupby(['author_ban_status']).median(numeric_only=True)[
    ['video_share_count']]
```

```python
# Creating a likes_per_view column
data['likes_per_view'] = data['video_like_count'] / data['video_view_count']

# Creating a comments_per_view column
data['comments_per_view'] = data['video_comment_count'] / data['video_view_count']

# Creating a shares_per_view column
data['shares_per_view'] = data['video_share_count'] / data['video_view_count']
```

```python
# compile the information in each of the three newly created columns for each combination of categories 
# of claim status and author ban status
data.groupby(['claim_status', 'author_ban_status']).agg(
    {'likes_per_view': ['count', 'mean', 'median'],
     'comments_per_view': ['count', 'mean', 'median'],
     'shares_per_view': ['count', 'mean', 'median']})
```

### Findings:
After looking at the dataset, the two variables that are most likely to help build a predictive model are claim_status and opinion_status. We can create a classification model to differentiate the two variables, as instructed from the employer.

The mean is larger than the median, so the data is right-skewed. Therefore, it is recommended to consider applying a variable transformation, such as taking logarithms, prior to modeling.
